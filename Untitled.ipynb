```python
import numpy as np
import pandas as pd
import os, sys
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv('/Users/kathanbhavsar/Downloads/parkinsons.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(15,8))
sns.heatmap(df.corr(),annot=True,cmap='viridis')
```

```python
df['status'].value_counts()
```

```python
X = df.drop(['status','name'],axis = 1)
y = df['status']
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = XGBClassifier()
```

```python
learning_rate = np.linspace(0,1,20)
n_estimators = [10,100,150,200,250]
param_grid = {'eta':learning_rate,'n_estimators':n_estimators}
from sklearn.model_selection import GridSearchCV
final_model = GridSearchCV(model,param_grid)
```

```python
final_model.fit(X_train_scaled,y_train)
```

```python
preds = final_model.predict(X_test)
```

```python
from sklearn.metrics import plot_confusion_matrix,classification_report
```

```python
print(classification_report(y_test,preds))
```
